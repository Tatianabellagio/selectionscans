```python

def summarize_temp(file_name):
    ds = xr.open_dataset(file_name)
    df = ds.to_dataframe()
    df = df.reset_index()
    ## convert temp to celcius
    df['t2m'] = df['t2m'] - 273.15
    ## convert temp to celcius
    site = file_name.split('_')[2]
    df['site'] = site
    df['month'] = df['time'].dt.month
    df['year'] = df['time'].dt.year
    df_max = df.groupby(['longitude', 'latitude','site','year', 'month'])['t2m'].max().reset_index().rename(columns={'t2m':'t2m_max'})
    df_min = df.groupby(['longitude', 'latitude','site','year', 'month'])['t2m'].min().reset_index().rename(columns={'t2m':'t2m_min'})
    df = pd.merge(df_max,df_min)
    return df

def summarize_prec_era5land(file_name):
    ds = xr.open_dataset(file_name)
    df = ds.to_dataframe().reset_index()
    ## convert total precipitation to mm
    df['tp_mm'] = df['tp'] * 1000
    site = file_name.split('_')[2]
    df['site'] = site
    df['day'] = df['time'].dt.day
    df['month'] = df['time'].dt.month
    df['year'] = df['time'].dt.year
    ## convert total precipitation to mm

    ## era5 land has data accumulated based on daily precipitation, so it accumulates over a day and then resets
    ## but it resets after 00:00:00, so the max of the previous day is shown at 00:00:00 of the next day 
    ## i will shift the column and take the max of each day (aka the accumulated in one day)
    df['tp_mm'] = df['tp_mm'].shift(-1)
    df = df.groupby(['longitude', 'latitude','site','year', 'month', 'day'])['tp_mm'].max().reset_index()
    ## then calcualte hte accumulated over the month 
    df = df.groupby(['longitude', 'latitude','site','year', 'month'])['tp_mm'].sum().reset_index()
    return df
```

```python
for site_1 in sites['site'].unique():
    for year in ['2018', '2019', '2020', '2021']:
        ## make sure it has all the prec and temp files
        #if all_files_psite[all_files_psite['site']==site_1]['nfiles'].values[0] >= 2:
            print(site_1, year)
            files_one_s = all_files_psite[(all_files_psite['site']==site_1) & (all_files_psite['year']==year)]['files'].values[0]
            
            files_one_s_prec = [i for i in files_one_s if 'prec' in i]
            ## get the temperature ones 
            files_one_s_temp = [i for i in files_one_s if 'temp' in i]
            
            all_df = {}
            for i in files_one_s_prec:
                df = summarize_prec_era5land('data/' + i)
                all_df[i] = df
            prec_df = pd.concat(all_df).reset_index(drop=True)
            
            all_df = {}
            for i in files_one_s_temp:
                df = summarize_temp('data/' + i)
                all_df[i] = df
            temp_df = pd.concat(all_df).reset_index(drop=True)
            
            climate_df = prec_df.merge(temp_df, on = ['longitude', 'latitude', 'site', 'year', 'month'])
            
            # Group by site and month, then calculate the average of each climate variable across all years
            climate_df_agg = climate_df.groupby(['site', 'month']).agg({
                't2m_max': 'mean',
                't2m_min': 'mean',
                'tp_mm': 'mean'
            }).reset_index()
            
            climate_df_agg.to_csv(f'climate_agg_files/climate_aggregate_site_{site_1}_year_{year}.csv',index=None)
```

```python
for year in ['2018', '2019', '2020', '2021']:
    files_year = [i for i in os.listdir('climate_agg_files') if 'site' in i and year in i]
    all_sites = {}
    for i in files_year:
        site = pd.read_csv('climate_agg_files/' + i)
        all_sites[i] = site
    
    all_sites = pd.concat(all_sites).reset_index(drop=True)
    
    print(all_sites['site'].nunique())
    
    climate_prec = all_sites.pivot(index='site', columns=['month'], values='tp_mm')
    climate_max_temp = all_sites.pivot(index='site', columns=['month'], values='t2m_max')
    climate_min_temp = all_sites.pivot(index='site', columns=['month'], values='t2m_min')
    
    climate_prec.to_csv(f'climate_prec_sites_year_{year}.csv')
    climate_max_temp.to_csv(f'climate_max_temp_sites_year_{year}.csv')
    climate_min_temp.to_csv(f'climate_min_temp_sites_year_{year}.csv')
    
    subprocess.call(f"~/miniforge3/envs/r-environment/bin/Rscript --vanilla /home/tbellagio/scratch/climate_grenet/run_dismo_bioclim_calc_sites.R {year}", 
                    shell=True)
```

```python
import rasterio
import pandas as pd

def extract_precipitation_for_sites(file_path, sites):
    # Read sites data from CSV
    # Open the GeoTIFF file with rasterio
    with rasterio.open(file_path) as src:
        # Initialize a list to store precipitation data
        precipitation_data = []
        
        # Loop through each site in the DataFrame
        for index, row in sites.iterrows():
            print(row)
            lat, long, site = row['latitude'], row['longitude'], row['site']
            print(lat)
            print(long)
            # Convert from geospatial coordinates to pixel coordinates
            col, row = src.index(long, lat)
            
            # Read the pixel value at the extracted coordinates
            data = src.read(1)  # 1 means we're reading from the first band
            value = data[row, col]
            
            # Append the site name and its precipitation value to the list
            print(value)
            precipitation_data.append({
                'site': site,
                'latitude': lat,
                'longitude': long,
                'precipitation': value
            })
    
    # Convert the list to a DataFrame
    return pd.DataFrame(precipitation_data)

# Path to your GeoTIFF file
file_path = 'wc2.1_30s_bio_12.tif'  # Update this path
sites = pd.read_csv('sites.csv').sample(10) # CSV containing sites with 'latitude', 'longitude', and 'site' columns

# Extract data
result_df = extract_precipitation_for_sites(file_path, sites)
print(result_df)
```

```python
import rasterio
from rasterio.transform import from_origin
from rasterio.warp import transform

def extract_data(file_path, lat, lon):
    with rasterio.open(file_path) as src:
        # Convert geographic coordinates (WGS84) to the coordinate system of the raster
        transformed_coord = transform(
            src.crs.geographic_crs, src.crs, [lon], [lat]
        )
        lon_proj, lat_proj = transformed_coord[0][0], transformed_coord[1][0]

        # Convert from projected coordinates to pixel coordinates
        col, row = src.index(lon_proj, lat_proj)
        
        # Ensure the row and column indices are within the bounds of the raster
        if not (0 <= row < src.height and 0 <= col < src.width):
            return None  # or handle the error as you see fit

        # Read the pixel value at the extracted coordinates
        data = src.read(1)  # 1 means we're reading from the first band
        value = data[row, col]
        
        return value

# Specify the path to your TIFF file and the coordinates
file_path = 'wc2.1_30s_bio_12.tif'  # Update this path
latitude =  48.541492  # Example latitude
longitude = 9.034098   # Example longitude

# Extract data
precipitation_value = extract_data(file_path, latitude, longitude)
if precipitation_value is not None:
    print(f'Annual Precipitation at the location: {precipitation_value} mm')
else:
    print("The specified coordinates are outside the bounds of the raster.")
```
