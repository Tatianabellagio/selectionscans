```python
import rasterio
from rasterio.transform import from_origin
import pandas as pd
import os

import pandas as pd
import rasterio
import geopandas as gpd
from shapely.geometry import Point
```

```python
sites = pd.read_csv('sites.csv')
```

```python
bios_old = pd.read_csv('../grene/data/worldclim_sitesdata.csv')[['site','bio1', 'bio2', 'bio3', 'bio4', 'bio5', 'bio6',
       'bio7', 'bio8', 'bio9', 'bio10', 'bio11', 'bio12', 'bio13', 'bio14',
       'bio15', 'bio16', 'bio17', 'bio18', 'bio19']]
```

```python
files = [i for i in os.listdir('wc/') if 'tif' in i]
```

```python
all_months = {}
for i in files:
    month = i.split('_')[-1].replace('.tif', '')
    # Open the TIFF file
    with rasterio.open('wc/' + i) as src:
        # Convert DataFrame to GeoDataFrame
        gdf = gpd.GeoDataFrame(sites, geometry=gpd.points_from_xy(sites.longitude, sites.latitude), crs='EPSG:4326')
        
        # Ensure the GeoDataFrame's CRS matches the raster's CRS
        gdf = gdf.to_crs(crs=src.crs)
    
        # Extract pixel values at the points' coordinates
        results = []
        for index, row in gdf.iterrows():
            for val in src.sample([(row.geometry.x, row.geometry.y)]):
                results.append((row['site'], val[0], row['longitude'], row['latitude']))
    
    # Create a results DataFrame
    results_df = pd.DataFrame(results, columns=['site', 'prec', 'longitude', 'latitude' ])
    all_months[month] = results_df
```

```python
prec = pd.concat(all_months).reset_index().drop('level_1', axis=1)
```

```python
prec.columns = ['month', 'site','prec', 'longitude', 'latitude']
```

```python
import rasterio
import geopandas as gpd
from shapely.geometry import Point
from rasterio.plot import show

# Path to the TIFF file (example with a bioclimatic variable)
file_path = 'wc2.1_30s_bio_12.tif'

# Longitude and latitude coordinates
longitude = 9.034098
latitude = 48.541492

# Create a GeoDataFrame with the point of interest
point = gpd.GeoDataFrame([{'geometry': Point(longitude, latitude)}], crs='EPSG:4326')

# Open the TIFF file
with rasterio.open(file_path) as src:
    # Ensure the point's CRS matches the raster's CRS
    point = point.to_crs(crs=src.crs)

    # Extract pixel values at the point's coordinates
    for val in src.sample([(point.geometry.x[0], point.geometry.y[0])]):
        print("The value at the specified location is:", val[0])

    # Optional: Show the raster
    show(src, title="WorldClim Data")
```

```python
import rasterio

# Longitude and latitude coordinates
longitude = 9.034098
latitude = 48.541492

# Path to the TIFF file
file_path = 'wc2.1_30s_bio_12.tif'

# Open the TIFF file
with rasterio.open(file_path) as src:
    # Check if the coordinate system of your points matches the dataset
    if src.crs != 'EPSG:4326':  # EPSG:4326 is the standard CRS for longitude and latitude
        print("Coordinate reference systems do not match. You might need to transform your coordinates.")

    # Convert geographic coordinates to pixel coordinates
    col, row = src.index(longitude, latitude)

    # Read the pixel value at the computed position
    value = src.read(1)[row, col]
    print("The value at location is:", value)
```
